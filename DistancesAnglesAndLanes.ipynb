```python
def project_position(x, y, vx, vy, dt=0.4):
    return x + vx*dt, y + vy*dt

def closest_point_on_segment(P, R, D):
    """
    Returns:
        s  -> distance along PR segment (scalar)
        closest_point -> coordinates of closest point
    """
    P = np.array(P)
    R = np.array(R)
    D = np.array(D)

    PR = R - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return 0, P

    PR_unit = PR / pass_length

    # projection length onto pass direction
    s = np.dot(D - P, PR_unit)

    # clamp to segment
    s = np.clip(s, 0, pass_length)

    closest = P + PR_unit * s

    return s, closest

def lane_obstructed(pass_row, tracking,
                    block_radius=4.0,
                    puck_speed=30.0):

    frame_id = pass_row["game_event_id"]
    poss_team = pass_row["event_team_id"]

    # Get defenders in this frame
    frame = tracking[tracking["game_event_id"] == frame_id]
    defenders = frame[frame["team_id"] != poss_team]

    # Passer and receiver positions
    P = np.array([
        pass_row["tracking_x_adj_i"],
        pass_row["tracking_y_adj_i"]
    ])

    R = np.array([
        pass_row["tracking_x_adj_j"],
        pass_row["tracking_y_adj_j"]
    ])

    PR = R - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return True

    PR_unit = PR / pass_length
    T_pass = pass_length / puck_speed

    for _, d in defenders.iterrows():

        D0 = np.array([
            d["tracking_x_adj"],
            d["tracking_y_adj"]
        ])

        Vd = np.array([
            d["tracking_vel_x_adj"],
            d["tracking_vel_y_adj"]
        ])

        # Find closest point along pass segment
        s, closest_point = closest_point_on_segment(P, R, D0)

        # Time puck reaches that closest point
        T = s / puck_speed

        # If puck already past receiver, skip
        if T > T_pass:
            continue

        # Puck position at time T
        puck_pos = P + PR_unit * s

        # Defender position at time T
        defender_pos = D0 + Vd * T

        # Check distance at same time
        if np.linalg.norm(defender_pos - puck_pos) < block_radius:
            return True

    return False
```

```python
pass_options["lane_blocked"] = pass_options.apply(
    lambda row: lane_obstructed(row, tracking),
    axis=1
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure
plt.figure(figsize=(10, 6))

# Separate teams
teams = t["team_name"].unique()

for team in teams:
    team_df = t[t["team_name"] == team]
    
    # Plot positions
    plt.scatter(
        team_df["tracking_x_adj"],
        team_df["tracking_y_adj"],
        label=team,
        s=100
    )
    
    # Plot velocity vectors
    plt.quiver(
        team_df["tracking_x_adj"],
        team_df["tracking_y_adj"],
        team_df["tracking_vel_x_adj"],
        team_df["tracking_vel_y_adj"],
        angles='xy',
        scale_units='xy',
        scale=1
    )
    
    # Add player labels
    for _, row in team_df.iterrows():
        if pd.notna(row["player_name"]):
            plt.text(
                row["tracking_x_adj"],
                row["tracking_y_adj"],
                row["player_name"],
                fontsize=8
            )

# Formatting
plt.axhline(0)
plt.axvline(0)
plt.gca().set_aspect('equal', 'box')
plt.xlabel("Adjusted X")
plt.ylabel("Adjusted Y")
plt.title("Player Positions (Adjusted Coordinates)")
plt.legend()
plt.show()
```
