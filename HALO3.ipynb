```python
def project_position(x, y, vx, vy, dt=0.3):
    return x + vx*dt, y + vy*dt

def closest_point_on_segment(P, R, D):
    """
    Returns:
        s  -> distance along PR segment (scalar)
        closest_point -> coordinates of closest point
    """
    P = np.array(P)
    R = np.array(R)
    D = np.array(D)

    PR = R - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return 0, P

    PR_unit = PR / pass_length

    # projection length onto pass direction
    s = np.dot(D - P, PR_unit)

    # clamp to segment
    s = np.clip(s, 0, pass_length)

    closest = P + PR_unit * s

    return s, closest

import numpy as np

import numpy as np

def lane_obstructed(pass_row, tracking,
                    cone_half_angle_deg=15.0,
                    puck_speed=30.0):

    frame_id = pass_row["game_event_id"]
    poss_team = pass_row["event_team_id"]

    frame = tracking[tracking["game_event_id"] == frame_id]
    defenders = frame[frame["team_id"] != poss_team]

    P = np.array([
        pass_row["tracking_x_adj_i"],
        pass_row["tracking_y_adj_i"]
    ])

    R = np.array([
        pass_row["tracking_x_adj_j"],
        pass_row["tracking_y_adj_j"]
    ])

    PR = R - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return False

    PR_unit = PR / pass_length
    tan_half_angle = np.tan(np.deg2rad(cone_half_angle_deg))

    T_pass_total = pass_length / puck_speed

    for _, d in defenders.iterrows():

        D0 = np.array([
            d["tracking_x_adj"],
            d["tracking_y_adj"]
        ])

        Vd = np.array([
            d["tracking_vel_x_adj"],
            d["tracking_vel_y_adj"]
        ])

        defender_speed = np.linalg.norm(Vd)

        # Closest point along pass
        s, closest = closest_point_on_segment(P, R, D0)

        # Skip if behind passer or beyond receiver
        if s <= 0 or s >= pass_length:
            continue

        # Check if that point is inside cone
        perp_dist = np.linalg.norm(D0 - closest)
        cone_radius_at_s = s * tan_half_angle

        inside_cone_now = perp_dist <= cone_radius_at_s

        # Time puck reaches this point
        T_puck = s / puck_speed

        # Defender arrival time to that point
        dist_to_point = np.linalg.norm(closest - D0)

        if defender_speed > 0:
            T_def = dist_to_point / defender_speed
        else:
            T_def = np.inf

        # BLOCK CONDITION
        if inside_cone_now or T_def <= T_puck:
            return True

    return False
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_passing_cone(P, R,
                      cone_angle_deg=12,
                      base_radius=1.5,
                      n_points=60):

    P = np.array(P)
    R = np.array(R)

    PR = R - P
    L = np.linalg.norm(PR)

    if L == 0:
        return

    PR_unit = PR / L
    theta = np.deg2rad(cone_angle_deg)

    # Perpendicular unit vector
    perp = np.array([-PR_unit[1], PR_unit[0]])

    s_vals = np.linspace(0, L, n_points)

    left_boundary = []
    right_boundary = []

    for s in s_vals:
        center = P + PR_unit * s
        radius = base_radius + s * np.tan(theta)

        left_boundary.append(center + perp * radius)
        right_boundary.append(center - perp * radius)

    left_boundary = np.array(left_boundary)
    right_boundary = np.array(right_boundary)

    # Plot center line
    plt.plot([P[0], R[0]], [P[1], R[1]],
             linestyle="--", linewidth=2)

    # Plot boundaries
    plt.plot(left_boundary[:, 0], left_boundary[:, 1], linewidth=1)
    plt.plot(right_boundary[:, 0], right_boundary[:, 1], linewidth=1)

    # Fill cone
    cone_polygon = np.vstack((left_boundary, right_boundary[::-1]))
    plt.fill(cone_polygon[:, 0], cone_polygon[:, 1], alpha=0.15)
```
