```python
def project_position(x, y, vx, vy, dt=0.3):
    return x + vx*dt, y + vy*dt

def closest_point_on_segment(P, R, D):
    """
    Returns:
        s  -> distance along PR segment (scalar)
        closest_point -> coordinates of closest point
    """
    P = np.array(P)
    R = np.array(R)
    D = np.array(D)

    PR = R - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return 0, P

    PR_unit = PR / pass_length

    # projection length onto pass direction
    s = np.dot(D - P, PR_unit)

    # clamp to segment
    s = np.clip(s, 0, pass_length)

    closest = P + PR_unit * s

    return s, closest

# Uses forward projection for path of pass
# If player reaches within 3 ft (reasonable reach) of puck path before puck passes them, they can intercept (counts as obstructed)

def lane_obstructed(pass_row, tracking,
                    base_angle=18.0,
                    min_angle=2,
                    alpha=0.03,
                    puck_speed=30.0,
                    intercept_radius=3,
                    dt=0.05):

    frame_id = pass_row["game_event_id"]
    poss_team = pass_row["event_team_id"]

    frame = tracking[tracking["game_event_id"] == frame_id]
    defenders = frame[frame["team_id"] != poss_team]

    P = np.array([
        pass_row["tracking_x_adj_i"],
        pass_row["tracking_y_adj_i"]
    ])

    R = np.array([
        pass_row["tracking_x_adj_j"],
        pass_row["tracking_y_adj_j"]
    ])

    Vr = np.array([
    pass_row["tracking_vel_x_adj_j"],
    pass_row["tracking_vel_y_adj_j"]
    ])

    PR = R - P
    pass_length = np.linalg.norm(PR)
    T_est = pass_length / puck_speed
    R_future = R + Vr * T_est
    
    PR = R_future - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return False

    PR_unit = PR / pass_length
    half_angle_deg = max(min_angle, base_angle * np.exp(-alpha * pass_length))
    half_angle_rad = np.deg2rad(half_angle_deg)
    tan_half_angle = np.tan(half_angle_rad)

    T_pass_total = pass_length / puck_speed

    for _, d in defenders.iterrows():

        D0 = np.array([
            d["tracking_x_adj"],
            d["tracking_y_adj"]
        ])

        Vd = np.array([
            d["tracking_vel_x_adj"],
            d["tracking_vel_y_adj"]
        ])

        # Check if currently inside cone
        s0, closest0 = closest_point_on_segment(P, R_future, D0)

        if 0 < s0 < pass_length:
            perp0 = np.linalg.norm(D0 - closest0)
            cone_radius0 = s0 * tan_half_angle
            if perp0 <= cone_radius0:
                return True

        # Time simulation
        t = 0.0
        while t <= T_pass_total:

            defender_pos = D0 + Vd * t

            s, closest = closest_point_on_segment(P, R_future, defender_pos)

            # Skip if defender is not between passer and receiver
            if s <= 0 or s >= pass_length:
                t += dt
                continue

            perp_dist = np.linalg.norm(defender_pos - closest)

            # Check if inside cone
            cone_radius = s * tan_half_angle
            inside_cone = perp_dist <= cone_radius

            # Check if within reach of puck
            within_intercept = perp_dist <= intercept_radius

            # Has puck passed the forechecker?
            puck_depth = puck_speed * t
            puck_not_passed = puck_depth <= s

            print("t:", round(t, 2),
              "s:", round(s, 2),
              "perp_dist:", round(perp_dist, 2),
              "within_intercept:", within_intercept,
              "puck_not_passed:", puck_not_passed)
            
            if within_intercept and puck_not_passed:
                return True

            t += dt

    return False
```

```python
# Changing receptions radius, uses forward projection for path of pass
# If player reaches within 3 ft (reasonable reach) of puck path before puck passes them, they can intercept (counts as obstructed)

def lane_obstructed2(pass_row, tracking,
                    puck_speed=30.0,
                    intercept_radius=3.0,
                    dt=0.05):

    frame_id = pass_row["game_event_id"]
    poss_team = pass_row["event_team_id"]

    frame = tracking[tracking["game_event_id"] == frame_id]
    defenders = frame[frame["team_id"] != poss_team]

    # Passer position
    P = np.array([
        pass_row["tracking_x_adj_i"],
        pass_row["tracking_y_adj_i"]
    ])

    # Receiver current position
    R = np.array([
        pass_row["tracking_x_adj_j"],
        pass_row["tracking_y_adj_j"]
    ])

    # Receiver velocity
    Vr = np.array([
        pass_row["tracking_vel_x_adj_j"],
        pass_row["tracking_vel_y_adj_j"]
    ])

    PR = R - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return False

    # Project receiver forward
    T_est = pass_length / puck_speed
    for _ in range(5):
        R_future = R + Vr * T_est
        T_est = np.linalg.norm(R_future - P) / puck_speed

    PR = R_future - P
    pass_length = np.linalg.norm(PR)

    if pass_length == 0:
        return False

    PR_unit = PR / pass_length
    T_pass_total = pass_length / puck_speed

    receiver_speed = np.linalg.norm(Vr)

    base_radius = 2.0  # minimum reception area
    r_rec = base_radius + 0.08 * pass_length # Increase radius as pass length increases

    # Prevent invalid arcsin
    ratio = min(0.99, r_rec / pass_length)

    half_angle_rad = np.arcsin(ratio)
    tan_half_angle = np.tan(half_angle_rad)

    # Defender location checks

    for _, d in defenders.iterrows():

        D0 = np.array([
            d["tracking_x_adj"],
            d["tracking_y_adj"]
        ])

        Vd = np.array([
            d["tracking_vel_x_adj"],
            d["tracking_vel_y_adj"]
        ])

        # Check current position 
        s0, closest0 = closest_point_on_segment(P, R_future, D0)

        if 0 < s0 < pass_length:
            perp0 = np.linalg.norm(D0 - closest0)
            cone_radius0 = s0 * tan_half_angle
            if perp0 <= cone_radius0:
                return True

        # Time simulation
        t = 0.0
        while t <= T_pass_total:

            defender_pos = D0 + Vd * t

            s, closest = closest_point_on_segment(P, R_future, defender_pos)

            if s <= 0 or s >= pass_length:
                t += dt
                continue

            perp_dist = np.linalg.norm(defender_pos - closest)

            # Inside tangent cone?
            cone_radius = s * tan_half_angle
            inside_cone = perp_dist <= cone_radius

            # Close enough to center line to physically intercept?
            within_intercept = perp_dist <= intercept_radius

            # Has puck reached this depth?
            puck_depth = puck_speed * t
            puck_not_passed = puck_depth <= s

            if inside_cone and within_intercept and puck_not_passed:
                return True

            t += dt

    return False
```

```python
def plot_passing_cone(
    P,
    R,
    Vr=None,
    puck_speed=30.0,
    base_radius=2.0,
    n_points=80
):

    P = np.array(P)
    R = np.array(R)

    # -------------------------------------------------
    # Project receiver forward (fixed-point iteration)
    # -------------------------------------------------
    if Vr is not None:
        Vr = np.array(Vr)

        PR0 = R - P
        L0 = np.linalg.norm(PR0)
        if L0 == 0:
            return

        T_est = L0 / puck_speed

        for _ in range(5):
            R_future = R + Vr * T_est
            T_est = np.linalg.norm(R_future - P) / puck_speed

        R = R_future

    # -------------------------------------------------
    # Recompute pass vector
    # -------------------------------------------------
    PR = R - P
    L = np.linalg.norm(PR)

    if L == 0:
        return

    PR_unit = PR / L
    T_pass_total = L / puck_speed
    pass_length = np.linalg.norm(PR)

    # -------------------------------------------------
    # Reception circle radius (same as lane_obstructed)
    # -------------------------------------------------
    if Vr is not None:
        receiver_speed = np.linalg.norm(Vr)
    else:
        receiver_speed = 0.0

    r_rec = base_radius + 0.08 * pass_length

    # Prevent invalid arcsin
    ratio = min(0.99, r_rec / L)

    half_angle_rad = np.arcsin(ratio)
    tan_half_angle = np.tan(half_angle_rad)

    # -------------------------------------------------
    # Construct cone from tangent geometry
    # -------------------------------------------------
    perp = np.array([-PR_unit[1], PR_unit[0]])
    s_vals = np.linspace(0, L, n_points)

    left_boundary = []
    right_boundary = []

    for s in s_vals:
        center = P + PR_unit * s
        radius = s * tan_half_angle

        left_boundary.append(center + perp * radius)
        right_boundary.append(center - perp * radius)

    left_boundary = np.array(left_boundary)
    right_boundary = np.array(right_boundary)

    # -------------------------------------------------
    # Plot center line
    # -------------------------------------------------
    plt.plot([P[0], R[0]], [P[1], R[1]],
             linestyle="--", linewidth=2)

    # Plot boundaries
    plt.plot(left_boundary[:, 0], left_boundary[:, 1], linewidth=1)
    plt.plot(right_boundary[:, 0], right_boundary[:, 1], linewidth=1)

    # Fill cone
    cone_polygon = np.vstack((left_boundary, right_boundary[::-1]))
    plt.fill(cone_polygon[:, 0], cone_polygon[:, 1], alpha=0.15)

    # -------------------------------------------------
    # Plot reception circle
    # -------------------------------------------------
    circle = plt.Circle(R, r_rec, fill=False, linestyle=":")
    plt.gca().add_patch(circle)
```
